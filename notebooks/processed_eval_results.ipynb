```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
steps = [i*100 for i in range(1, 11)]
```

```python
with open("../processed_eval_results.pkl", "rb") as f:
    results = pickle.load(f)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# plot all erdos experiments
axs[0, 0].plot(steps, results["erdos_pagerank"], label="erdos_pagerank")
axs[0, 0].plot(steps, results["erdos_degree"], label="erdos_degree")
axs[0, 0].plot(steps, results["erdos_betweeness"], label="erdos_betweeness")
axs[0, 0].set(xlabel="games played", ylabel="mean rewards")
axs[0, 0].set_title("Network: Erdos-Renyi")
axs[0, 0].legend()

# plot all barabasi experiments
axs[0, 1].plot(steps, results["barabasi_pagerank"], label="barabasi_pagerank")
axs[0, 1].plot(steps, results["barabasi_degree"], label="barabasi_degree")
axs[0, 1].plot(steps, results["barabasi_betweeness"], label="barabasi_betweeness")
axs[0, 1].set(xlabel="games played", ylabel="mean rewards")
axs[0, 1].set_title("Network: Barabasi")
axs[0, 1].legend()

# plot all watts experiments
axs[0, 2].plot(steps, results["watts_pagerank"], label="watts_pagerank")
axs[0, 2].plot(steps, results["watts_degree"], label="watts_degree")
axs[0, 2].plot(steps, results["watts_betweeness"], label="watts_betweeness")
axs[0, 2].set(xlabel="games played", ylabel="mean rewards")
axs[0, 2].set_title("Newtwork: Watts")
axs[0, 2].legend()

# plot all pagerank experiments
axs[1, 0].plot(steps, results["erdos_pagerank"], label="erdos_pagerank")
axs[1, 0].plot(steps, results["barabasi_pagerank"], label="barabasi_pagerank")
axs[1, 0].plot(steps, results["watts_pagerank"], label="watts_pagerank")
axs[1, 0].set(xlabel="games played", ylabel="mean rewards")
axs[1, 0].set_title("Centrality: Pagerank")
axs[1, 0].legend()

# plot all degree experiments
axs[1, 1].plot(steps, results["erdos_degree"], label="erdos_degree")
axs[1, 1].plot(steps, results["barabasi_degree"], label="barabasi_degree")
axs[1, 1].plot(steps, results["watts_degree"], label="barabasi_betweeness")
axs[1, 1].set(xlabel="games played", ylabel="mean rewards")
axs[1, 1].set_title("Centrality: Degree")
axs[1, 1].legend()

# plot all betweeness experiments
axs[1, 2].plot(steps, results["erdos_betweeness"], label="erdos_betweeness")
axs[1, 2].plot(steps, results["barabasi_betweeness"], label="barabasi_betweeness")
axs[1, 2].plot(steps, results["watts_betweeness"], label="watts_betweeness")
axs[1, 2].set(xlabel="games played", ylabel="mean rewards")
axs[1, 2].set_title("Centrality: Degree")
axs[1, 2].legend()

plt.show()
```
